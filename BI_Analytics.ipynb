```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
from datasets import Dataset
import json
import pandas as pd
```

```python
df = pd.read_csv('drive/MyDrive/Colab Notebooks/Cathys_test_IZ/dataset.csv')
df.info()
```

```python

# Format and Conversion
# Convert date columns to datetime, coercing errors
date_columns = ['order_date_time', 'Issue_reported at', 'issue_responded', 'Survey_response_Date']
for col in date_columns:
    df[col] = pd.to_datetime(df[col], format='%d/%m/%Y %H:%M', errors='coerce')

# Convert Item_price to numeric, coercing errors
df['Item_price'] = pd.to_numeric(df['Item_price'], errors='coerce')

display(df.head(2))
```

```python


average_csat = df['CSAT Score'].mean()

average_issue_resolution_time = (df['issue_responded'] - df['Issue_reported at']).mean()
average_handling_time = df['connected_handling_time'].mean()


# Analyze CSAT score distribution
csat_distribution = df['CSAT Score'].value_counts().sort_index(ascending=False)


csat_by_channel = df.groupby('channel_name')['CSAT Score'].mean().sort_values(ascending=False)
csat_by_category = df.groupby('category')['CSAT Score'].mean().sort_values(ascending=False)
csat_by_sub_category = df.groupby(['category', 'Sub-category'])['CSAT Score'].mean().sort_values(ascending=False)


csat_by_agent = df.groupby('Agent_name')['CSAT Score'].mean().sort_values(ascending=False)
csat_by_supervisor = df.groupby('Supervisor')['CSAT Score'].mean().sort_values(ascending=False)
csat_by_manager = df.groupby('Manager')['CSAT Score'].mean().sort_values(ascending=False)


csat_by_tenure = df.groupby('Tenure Bucket')['CSAT Score'].mean().sort_values(ascending=False)
csat_by_shift = df.groupby('Agent Shift')['CSAT Score'].mean().sort_values(ascending=False)

# print(f"2. Average Issue Resolution Time: {issue_resolution_time}")
# print(f"3. Average Handling Time: {average_handling_time:.2f} minutes")


display(f"--- Overall Average CSAT Score: {average_csat:.2f} ---") # Format to 2 decimal places

display(f"--- Average Issue Resolution Time: {average_issue_resolution_time} ---")

display(f"--- Average Handling Time: {average_handling_time:.2f} minutes ---")


display("\n--- CSAT Score Distribution ---")
display(csat_distribution)

display("\n--- CSAT by Channel ---")
display(csat_by_channel)

display("\n--- CSAT by Category ---")
display(csat_by_category)

display("\n--- CSAT by Sub-category ---")
display(csat_by_sub_category)

display("\n--- CSAT by Agent (top performing) ---")
display(csat_by_agent.head())

display("\n--- CSAT by Supervisor (top performing) ---")
display(csat_by_supervisor.head())

display("\n--- CSAT by Manager (top performing) ---")
display(csat_by_manager.head())

display("\n--- CSAT by Tenure Bucket ---")
display(csat_by_tenure)

display("\n--- CSAT by Agent Shift ---")
display(csat_by_shift)


# # Calculate key metrics
# average_csat = df['CSAT Score'].mean()
# csat_by_channel = df.groupby('channel_name')['CSAT Score'].mean()
# issue_resolution_time = (df['issue_responded'] - df['Issue_reported at']).mean()
# average_handling_time = df['connected_handling_time'].mean()
# csat_by_category = df.groupby('category')['CSAT Score'].mean()


# print("--- Key Performance Metrics ---")
# print(f"1. Overall Average CSAT Score: {average_csat:.2f}")
# print(f"2. Average Issue Resolution Time: {issue_resolution_time}")
# print(f"3. Average Handling Time: {average_handling_time:.2f} minutes")
# print(f"4. CSAT by Channel:\n{csat_by_channel}")
# print(f"5. CSAT by Category (Top 5):\n{csat_by_category.sort_values(ascending=False).head()}")
```
